Now let's try to read the roads dataset for Cuyahoga county from Census Bureau

Finally, we look at the vrious counties in Ohio using the county boundary data from Ohio Department of Transportation (ODOT).

```python
ohCounties= gpd.read_file(r'../../largedatasets/ohio_county_boundaries/ODOT_County_Boundaries.shp')
ohCounties
```

This dataset has 88 rows (as there are 88 counties in Ohio) and 24 columns (geometry being one of them). The geometry column data indicates that the spatial model associated with dataset is vector and the type is Polygon.

You can perform any Pandas type operation with the non-geometry columns of GeoDataFrame. For example lets sort the GeoDataFrame based on the POP_2010 column

```python
ohCounties.sort_values(by='POP_2010',ascending=False)
```

You can see that Cuyahoga is the county with the highest population. We can easily caclulate the total population based on POP_2010.

```python
ohCounties['POP_2010'].sum()
```

Around 11.5 million. Let's calculate the change in population between 2010 and 2000

```python
ohCounties['POP_CHANGE_2010_2000'] = ohCounties['POP_2010'] - ohCounties['POP_2000']
ohCounties
```
